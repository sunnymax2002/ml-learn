# Study of Economic Inequality

This is an attempt to analyze the Economic Inequality in different world regions, and extract insight into the underlying reasons for the same, and possible solutions to associated problems

We first start with importing the dataset from https://wid.world/, and extract one economic variable's data and plot it, for illustration

```python
import pandas as pd
from os import path
import matplotlib.pyplot as plt

dir_wid_dataset = r"D:\Learning\Datasets\wid_all_data_20200615"

# Country = INDIA
country_ID = "IN"

# Open country metadata and data as Pandas Dataframe(s): Use 'variable' column as index
file_metadata = path.join(dir_wid_dataset, "WID_metadata_" + country_ID + ".csv")
df_metadata = pd.read_csv(file_metadata, sep=';', index_col='variable')
print("\nMetadata Info:")
df_metadata.info()

file_data = path.join(dir_wid_dataset, "WID_data_" + country_ID + ".csv")
df_data = pd.read_csv(file_data, sep=';', index_col='variable')
print("\nData Info:")
df_data.info()

# Specify the economic analysis / variable
# Variable values don't match: metadata has i before 99x, while data has i after 99x...
dvar_val = 'mgdpro999i'
mvar_val = 'mgdproi999'

# SELECT * FROM DF_DATA WHERE INDEX=var_val
econ_var_desc = df_metadata.loc[mvar_val]['simpledes']
print("\nSelected Economic Data (" + dvar_val + "): " + econ_var_desc + "\n")
df_econ_data = df_data.loc[dvar_val]
print(df_econ_data)

fig, ax = plt.subplots()
ax.plot(df_econ_data['year'], df_econ_data['value'])
plt.show()
```
